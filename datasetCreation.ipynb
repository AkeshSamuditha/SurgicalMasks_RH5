# Dataset 1

```python
health_data = pd.read_csv('healthcare_dataset.csv')
health_data = health_data.drop(['Insurance Provider', 'Room Number', 'Name', 'Billing Amount'], axis=1)

# convert date to datetime
health_data.head()
```

```python
# create length of stay column
health_data['Date of Admission'] = pd.to_datetime(health_data['Date of Admission'])
health_data['Discharge Date'] = pd.to_datetime(health_data['Discharge Date'])
health_data['Length of Stay'] = health_data['Discharge Date'] - health_data['Date of Admission']
health_data['Length of Stay'] = health_data['Length of Stay'].dt.days
health_data.drop(['Date of Admission'], axis=1, inplace=True)

health_data['Medical Condition'] = health_data['Medical Condition'].replace('Arthritis', 'Leptospirosis')
health_data['Medical Condition'] = health_data['Medical Condition'].replace('Cancer', 'CKD')
health_data['Medical Condition'] = health_data['Medical Condition'].replace('Obesity', 'Malnutrition')

health_data.head()
```

```python
import random
asthma_count = health_data['Medical Condition'].str.contains('Asthma', case=False).sum()
replace_count = int(0.7* asthma_count)
asthma_indices = health_data[health_data['Medical Condition'].str.contains('Asthma', case=False)].index
replace_indices = random.sample(list(asthma_indices), replace_count)
health_data.loc[replace_indices, 'Medical Condition'] = 'Tuberculosis'

hypertension_count = health_data['Medical Condition'].str.contains('Hypertension', case=False).sum()
replace_count = int(0.4* hypertension_count)
Hypertension_indices = health_data[health_data['Medical Condition'].str.contains('Hypertension', case=False)].index
replace_indices = random.sample(list(Hypertension_indices), replace_count)
health_data.loc[replace_indices, 'Medical Condition'] = 'Dengue'

count = health_data['Medical Condition'].str.contains('Hypertension', case=False).sum()
replace_count = int(0.6*count)
indices = health_data[health_data['Medical Condition'].str.contains('Hypertension', case=False)].index
replace_indices = random.sample(list(indices), replace_count)
health_data.loc[replace_indices, 'Medical Condition'] = 'Pneumonia'


count = health_data['Medical Condition'].str.contains('Malnutrition', case=False).sum()
replace_count = int(0.2*count)
indices = health_data[health_data['Medical Condition'].str.contains('Malnutrition', case=False)].index
replace_indices = random.sample(list(indices), replace_count)
health_data.loc[replace_indices, 'Medical Condition'] = 'Leptospirosis'

count = health_data['Medical Condition'].str.contains('Diabetes', case=False).sum()
replace_count = int(0.4*count)
indices = health_data[health_data['Medical Condition'].str.contains('Diabetes', case=False)].index
replace_indices = random.sample(list(indices), replace_count)
health_data.loc[replace_indices, 'Medical Condition'] = 'CKD'

count = health_data['Medical Condition'].str.contains('Leptospirosis', case=False).sum()
replace_count = int(0.5*count)
indices = health_data[health_data['Medical Condition'].str.contains('Leptospirosis', case=False)].index
replace_indices = random.sample(list(indices), replace_count)
health_data.loc[replace_indices, 'Medical Condition'] = 'Renal Stones'

count = health_data['Medical Condition'].str.contains('Malnutrition', case=False).sum()
replace_count = int(0.4*count)
indices = health_data[health_data['Medical Condition'].str.contains('Malnutrition', case=False)].index
replace_indices = random.sample(list(indices), replace_count)
health_data.loc[replace_indices, 'Medical Condition'] = 'Renal Stones'
```
